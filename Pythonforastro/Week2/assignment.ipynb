```python
#INTERMEDIATE

import matplotlib.pyplot as plt
import numpy as np

image1=[[0,12 ,11 ,9],
[2 ,59 ,98 ,11],
[8 ,75 ,67 ,12],
[4 ,13 ,15 ,3]]

image2=[[0, 11, 11, 7,],
[0 ,57 ,94 ,10],
[3 ,75 ,62 ,8],
[1 ,13 ,13 ,1]]

image3=[[1, 12, 11, 11],
[0, 55, 92, 10],
[2, 73, 62, 11],
[5, 12, 12, 2]]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

heatmap1 = axes[0].matshow(image1, cmap='viridis')
heatmap2 = axes[1].matshow(image2, cmap='plasma')
heatmap3 = axes[2].matshow(image3, cmap='inferno')

colorbar1 = plt.colorbar(heatmap1, ax=axes[0])
colorbar2 = plt.colorbar(heatmap2, ax=axes[1])
colorbar3 = plt.colorbar(heatmap3, ax=axes[2])

plt.tight_layout()

plt.show()
```

```python
#ADVANCED

import pandas as pd
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(15, 5))

data1 = pd.read_csv("ngc_1068.csv")
data2 = pd.read_csv("3c273.csv")
data3 = pd.read_csv("ngc_1068.csv")
data4 = pd.read_csv("ngc_4565.csv")

heatmap1 = axes[0].matshow(data1)
heatmap2 = axes[1].matshow(data2)
heatmap3 = axes[2].matshow(data3)
heatmap4 = axes[3].matshow(data4)

colorbar1 = plt.colorbar(heatmap1, ax=axes[0])
colorbar2 = plt.colorbar(heatmap2, ax=axes[1])
colorbar3 = plt.colorbar(heatmap3, ax=axes[2])
colorbar4 = plt.colorbar(heatmap4, ax=axes[3])

plt.tight_layout()
plt.show()
```

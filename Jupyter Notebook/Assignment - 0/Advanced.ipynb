```python
# 1. Placing points on graph

import matplotlib.pyplot as plt

x = [1, 2, 3, 5, 6]
y = [2, 4, 5, 2, 1]

plt.plot(x, y, 'bo')

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Plotting Points')

plt.show()
```

```python
# 2. connecting points on graph

import matplotlib.pyplot as plt

x = [1, 2, 3, 5, 6]
y = [2, 4, 5, 2, 1]

plt.plot(x, y, 'bo-')

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Connecting Points')

plt.show()
```

```python
# 3. plotting a function

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)

y = 3 * x + 1

plt.plot(x, y)

# Set the axis labels
plt.xlabel('X')
plt.ylabel('Y')

# Set the title of the graph
plt.title('Plotting Function y = 3x + 1')

# Display the graph
plt.show()
```

```python
# 4. plotting sin(x) on graph

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

y = np.sin(x)

plt.plot(x, y)

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Plotting Function y = sin(x)')

plt.show()
```

```python
# 5. plotting sin(x)+cos(x)

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

y = np.sin(x) + np.cos(x)

plt.plot(x, y)

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Plotting Function y = sin(x)+cos(x)')

plt.show()
```

```python
# 6. plotting sin(x) and 3sin(x) in the same graph and spotting the differences

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

y1 = np.sin(x)
y2 = 3 * np.sin(x)

plt.plot(x, y1, label='sin(x)')

plt.plot(x, y2, label='3sin(x)', color='red')

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Plotting Sine Functions')

plt.legend()

plt.show()
```

```python
# 7. plotting sin(x) and sin(x+30) in the same graph and spotting the difference

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

y1 = np.sin(x)
y2 = np.sin(x + np.deg2rad(30))

plt.plot(x, y1, label='sin(x)')

plt.plot(x, y2, label='sin(x + 30)', color='red')

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Plotting Functions')

plt.legend()

plt.show()

```

```python
# 8. plotting 2d array using matshow function

import numpy as np
import matplotlib.pyplot as plt

arr = np.array([[2, 3, 1], [1, 23, 3], [0, 2, 1]])

plt.matshow(arr)

plt.colorbar()

plt.title('Plotting 2D Array')

plt.show()
```

```python
# 9. plotting another 2d array using matshow function

import numpy as np
import matplotlib.pyplot as plt

array = np.zeros((20, 20))

np.fill_diagonal(array, 100)

plt.matshow(array)

plt.colorbar()

plt.title('Plotting 2D Array')

plt.show()
```

```python
# 11. measuring voltage and current through resistor

import numpy as np
import matplotlib.pyplot as plt

voltage = np.array([10, 9, 9.5, 12, 14, 15])
current = np.array([5.25, 4.25, 4.50, 6.25, 7.25, 7.50])

plt.scatter(voltage, current, label='Measured Data')

resistance = voltage / current
mean_resistance = np.mean(resistance)

fit_line = np.polyfit(voltage, current, 1)
best_fit_line = np.poly1d(fit_line)

plt.plot(voltage, best_fit_line(voltage), label='Best Fit Line')

plt.xlabel('Applied Voltage (V)')
plt.ylabel('Measured Current (mA)')

plt.title('Voltage vs Current')

plt.legend()

plt.show()

residuals = current - best_fit_line(voltage)

plt.scatter(voltage, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Applied Voltage (V)')
plt.ylabel('Residuals')
plt.title('Residuals')
plt.show()

print('Equation of the best fit line: y =', fit_line[0], 'x +', fit_line[1])
print('Resistance (mean value):', mean_resistance)
```
